# Multilayer Perceptron

Using MLP (from keras) to recognize MNIST chars.

```python
import numpy as np
from keras.datasets import mnist
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data() #carrega os conjuntos de treinamento (_train) e teste (_test)

X_train = X_train.reshape(60000, 784).astype('float32') / 255
X_test = X_test.reshape(10000, 784).astype('float32') / 255
X_train.shape
```

```python
from keras.utils import to_categorical

Y_train = to_categorical(Y_train, num_classes=10)
Y_test = to_categorical(Y_test, num_classes=10)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
model = Sequential()

model.add(Dense(64, input_dim=784))
model.add(Activation('relu'))

model.add(Dense(32))
model.add(Activation('relu'))

model.add(Dense(10))
model.add(Activation('softmax'))

sgd = SGD(learning_rate=0.01, momentum=0.99, nesterov=True)
model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train,
                    Y_train,
                    epochs=5,
                    batch_size=256,
                    validation_data=(X_test, Y_test))
```
