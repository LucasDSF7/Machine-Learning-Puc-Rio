# Multilayer Perceptron

Using MLP (using only numpy) to recognize MNIST chars.

## Importing Data

MNIST dataset has n row with 784 columns, each column representing a pixel from a 28x28 image of a digit.
We first need to transpose this dataset, to get 784 rows with n columns to feed the algorithm.

## Undestanding layers

In each layer, each neuron receives a weighted input value with a bias, then its applyied a activation function.

    Z¹ = W¹A° + B¹
    A¹ = Activation(Z¹)

In this case A¹ would be like a input for the next layer.

    W (nº neurons next layer x nº neurons layer before (or nº features))
    A (nº neurons layer before (or nº features) x training size)
